#### Exploring hashtags included in the tweets

```python
# Create a new DataFrame containing only the text.
tweets_text = pd.DataFrame(tweets['tweet_full_text'])

# View the DataFrame.
tweets_text.head()
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
# Display the first 30 records.
tseries = pd.Series(tags).value_counts()
tseries.head(30)
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
data = pd.DataFrame(tseries).reset_index()

# View dataframe.
data.head()
```

```python
# Rename columns
data.columns = ['word', 'count']

# Check column names updated
data.head()
```

```python
# Check data types of columns in dataframe
data.info()
```

```python
# Display records where the count is larger than 10.
tag10 = data[data['count']>10]

# Check filter applied correctly.
tag10
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.
sns.barplot(x= 'count', y= 'word',estimator= sum, data= tag10)

# Improve chart by adding clear labels
plt.xlabel('Frequency of use',fontsize=14)
plt.ylabel('Word', fontsize=14)
plt.title('Hashtags by popularity',fontsize=20)
```

#### Above visualisation could be improved by removing the hastag 'healthcare' which is over represented and by reducing the number of hashtags to show a summary

```python
# Create the plot to remove over represented hashtag and provide a summary.
sns.barplot(x= 'count', y= 'word',estimator= sum, data= tag10.loc[(tag10['count'] >15) & (tag10['count'] <100)])

# Improve chart by adding clear labels
plt.xlabel('Frequency of use',fontsize=14)
plt.ylabel('Word', fontsize=14)
plt.title('Hashtags by popularity summary',fontsize=20)
```
